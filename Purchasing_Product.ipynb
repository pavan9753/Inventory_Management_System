```python
import json
import time
import math
import emoji

while(True):
  print("**********************************")
  print("\U0001F64F",end="") 
  print("\U0001F64F",end="") 
  print("\U0001F64F",end="") 
  print("\U0001F64F",end="") 
  print("\U0001F64F",end="")
  print("\U0001F64F")
  print("Welcome to the ETG Shopping Mall ")
  print("**********************************")
  s=0
  cart={}
  pid=[]
  pname=[]
  pprice=[]
  pquantity=[]
  pbill=[]
  ptime=[]
  pmfd=[]
  while(True):
    
    print("\n1.Purchasing a product ")
    print("2.My Current Billing Amount ")
    print("3.Display My products in cart ")
    print("4.Total Number of products in my Cart ")
    print("5.Remove item from my cart ")
    print("6.Product with Maximum Billing Amount ")
    print("7.Product with Minimum Billing Amount ")
    print("8.Exit")
    y=int(input("Choose one of the service from the MENU given :\n"))
    if(y==1):
      id=input("Enter the product id :")
      if(id in records.keys()):
        print("Quantity Available in the Store : ",records[id]['quantity'])
        q=int(input("enter the number of items to be purchased of this type: "))
        if(q<=records[id]['quantity']):
          print(emoji.emojize(":grinning_face_with_big_eyes:"),end="")
          print(emoji.emojize(":grinning_face_with_big_eyes:"),end="")
          print(emoji.emojize(":grinning_face_with_big_eyes:"),end="")
          print(emoji.emojize(":grinning_face_with_big_eyes:"),end="")
          print(emoji.emojize(":grinning_face_with_big_eyes:"),end="")
          print(emoji.emojize(":grinning_face_with_big_eyes:"))
          print("You have succesfully purchased ",q," number of items of ",records[id]['name'])
          #Adding the items to the corresponding lists
          pid.append(id)
          pname.append(records[id]['name'])
          pprice.append(records[id]['price'])
          pquantity.append(q)
          pbill.append(records[id]['price']*q)
          ptime.append(time.ctime())
          pmfd.append(records[id]['mfd'])
          records[id]['quantity']-=q   
          #printing the details of the product choosen by customer
          print('****************')
          cart[id]={'name':records[id]['name'],'quantity':q,'price':records[id]['price'],'bill':records[id]['price']*q,'time':time.ctime()}
          print('product purchased : ',records[id]['name'])
          print('product id : ',id)
          print('expiry date : ',records[id]['exp'])  
          print('time : ',time.ctime())
          print('quantity : ',q)
          print('*************************************')
          print('total billing for this product :',records[id]['price']*q) 
          s=s+records[id]['price']*q
          print('*************************************\n') 
          #print(pname,pprice,pquantity,pbill,ptime,pmfd)
        else:
          print("not sufficient ,we have only ",records[id]['quantity']," number of quantities ")
      else:
        print("Sorry!!! No Product id have been matched with your product id")
    elif(y==2):
      # total amount till now
      print("****************************")
      print("My current Billing : ",s," rupees")
      print("****************************\n")
    elif(y==3):
      x=1
      if(len(cart)==0):
        print("*********  SORRY!!!!!!  ***********")
        print("No Items in your cart to display\n")
        continue
      print("The Items in the Cart are :")
      for i in cart.keys():
        print("***********************")
        print("Item ",x," in cart")
        x+=1
        print("Product id : ",i)
        print("product name : ",cart[i]['name'])
        print("Quantity taken : ",cart[i]['quantity'])
        print("Price of one product : ",cart[i]['price'])
        print("total price for this product [price*quantity] : ",cart[i]['bill'])
        print("Time of purchase : ",cart[i]['time'])
        print("***********************")
      x=0
    elif(y==4):
      #length of the cart
      print("********************")
      print("Total number of products in my Cart : ",len(cart))
      print("********************")
    elif(y==5):
      x=input("Are you sure ,Do you want to delete it [Y/N] ")
      if(x=='Y'):
        te=input("Enter the product id for which product to be removed : ")
        if(te not in cart.keys()):
          print("No such product with the given product id")
        else:
          print("\U0001F625",end="") 
          print("\U0001F625",end="") 
          print("\U0001F625",end="") 
          print("\U0001F625",end="") 
          print("\U0001F625",end="") 
          print("\U0001F625")
          print("Name of the product : ",cart[te]['name'])
          print("Successfully deleted!! ",cart[te]['name']," from the cart")
          pname.pop(pid.index(te))
          pprice.pop(pid.index(te))
          pquantity.pop(pid.index(te))
          ptime.pop(pid.index(te))
          pbill.pop(pid.index(te))
          pmfd.pop(pid.index(te))
          pid.remove(te)
          del cart[te]
          #print(len(cart))
      else:
        print("Item not deleted !!")
        continue

    elif(y==6):
      mm=-9999
      for i in cart.keys():
        mm=max(cart[i]['bill'],mm)
      print("************************************")
      print("\nThe product with Maximum billing Amount in the Cart : ",mm)
      print("************************************")

    elif(y==7):
      mm=99999999
      for i in cart.keys():
        mm=min(cart[i]['bill'],mm)
      print("************************************")
      print("\nThe product with Minimum billing Amount in the Cart : ",mm)
      print("************************************")

    else:
      
      
      print("THANK YOU!! ")
      print("\U0001F64F",end="") 
      print("\U0001F64F",end="") 
      print("\U0001F64F",end="") 
      print("\U0001F64F",end="") 
      print("\U0001F64F",end="") 
      print("\U0001F64F")
      print("VISIT AGAIN!")
      #print(pname,pprice,pquantity,pbill,ptime,pmfd)
      #print(len(cart))
      if(len(cart)!=0):
        sales[len(sales)+1]={'id':pid,'name':pname,'price':pprice,'quantity':pquantity,'bill':pbill,'time':ptime,'mfd':pmfd,'total_amount':s}
      js=json.dumps(sales)
      fd=open("sales.json","w")
      fd.write(js)
      fd.close()
      break
  var=input("Are there still any customer who wants to visit the Shopping mall : [Y/N] ")
  if(var=='N'):
    break
    
    

```

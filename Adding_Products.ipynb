# IMPORTING RECORDS FILE


```python
import json as js
fd=open("records.json","r")
text=fd.read()
fd.close()
records=js.loads(text)
records
```

# ADDING NEW PRODUCTS TO RECORDS

```python
tot=int(input("Enter the total number of products to be added to list "))
print("\n********##########*********")
for i in range(tot):
  print("Enter the ",i+1,"th product details:")
  print("****************************\n")
  pid=input("Enter the product id :")
  pname=input("Enter the product name :")
  p=int(input("Enter the product price :"))
  q=int(input("Enter the product quantity :"))
  mfd=input("enter the manufacturing date of product [DD/MM/YYYY] : ")
  exp=input("enter the expiry date of product [DD/MM/YYYY] : \n")
  records[pid]={'name':pname,'price':p,'quantity':q,'mfd':mfd,'exp':exp}
j=js.dumps(records)
fd=open("records.json","w")
fd.write(j)
fd.close()
```

# Updating the specific product ID



```python
while(True):
  id=input("Enter the new product id whose quantity to be updated :")
  print("****************************")
  print("Name of the product : ",records[id]['name'])
  print("Price of a single piece of product : ",records[id]['price'])
  print("Before adding the product's quantity is :",records[id]['quantity'])
  q=int(input("Enter the quantity to be added :"))
  records[id]['quantity']=records[id]['quantity']+q
  print("Successfully updated")
  print("Product name is :",records[id]['name'])
  print("After adding the product's quantity is :",records[id]['quantity'],"\n")
  temp=input("Do you want to continue : [Y/N] ")
  if(temp=='N'):
    break
```
